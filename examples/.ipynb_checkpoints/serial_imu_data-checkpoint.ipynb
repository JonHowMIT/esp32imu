# Read Encoder from Motor via Serial

```python
from os.path import exists
file_exists = exists('basic_material.py')

if (not file_exists):
  !wget -O basic_material.py https://www.dropbox.com/s/o48uhuwah8n2xww/basic_material.py?dl=0
%run basic_material

try: 
    import serial
except:
    !pip3 install pyserial
    import serial

import time, csv, os, requests
plt.rcParams["figure.figsize"] = (8,5)
 
from os import path
if os.path.isdir("./data/"):
    pass
else:
    !mkdir ./data
```

```python
port = '/dev/cu.usbserial-1110'
ser = serial.Serial(port,baudrate=115200)
ser.flushInput()

# sum will treat Boolean True as a 1, False as a 0
n = sum(os.path.isfile(name) for name in os.listdir('.'))
file_name = "Encoder_data_%s.cvs" % n 
print("Storing recorded data in",file_name)
```

```python
# Initialize 
Tf = 20
Gyro = []
mAcc = []
t = []
count = []

Reading_data = False
File_end = False

ser.write(bytearray(b'R'))
time.sleep(1)
print('Start Moving Cube')
ser.write(bytearray(b'C'))
time.sleep(Tf)
ser.write(bytearray(b'S'))
print('Stop Moving Cube')

while not File_end:
    ser_byte = ser.readline()
    try:
        decoded_bytes = ser_byte.decode('ascii').split(',')
        if (float(decoded_bytes[0]) > 0) and (Reading_data == False):
            print('Collecting Data')
            Reading_data = True
    except:
        pass
    if Reading_data:
        try:
            if (float(decoded_bytes[1]) > Tf-2):
                File_end = True
            dc = float(decoded_bytes[0])
            dt = float(decoded_bytes[1])
            s = np.copy(decoded_bytes[2:5])
            da = [float(value) for value in s]
            ss = np.copy(decoded_bytes[5:8])
            dg = [float(value) for value in ss]
            count.append(dc) 
            t.append(dt) 
            mAcc.append([da[0], da[1], da[2]])
            Gyro.append([dg[0], -dg[1], dg[2]])
            with open(file_name, "a+", newline='') as file:
                writer = csv.writer(file, delimiter=',')
                writer.writerow([dt, da[0], da[1], da[2], dg[0], -dg[1], dg[2]])
        except:
            print(ser_byte)
```

```python
t = np.array(t)
t -= t[0]    
mAcc = np.array(mAcc)
Gyro = np.array(Gyro)

lt = len(t)
la = min(len(mAcc),lt)
lg = min(len(Gyro),lt)
```

```python
# parts to plot
fig, ax = plt.subplots()
AccX = mAcc.flatten()[0::3]
AccY = mAcc.flatten()[1::3]
AccZ = mAcc.flatten()[2::3]
ax.plot(t[:la],AccX[:la],'b',label='Ax')
ax.plot(t[:la],AccY[:la],'r',label='Ay')
ax.plot(t[:la],AccZ[:la],'g',label='Az')
#ax.set_ylim([-1.5, 1.5])
plt.legend()
plt.show()

fig, ax = plt.subplots()
ax.plot(t[:lg],Gyro.flatten()[0:3*lg:3],'b',label='Gx')
ax.plot(t[:lg],Gyro.flatten()[1:3*lg:3],'r',label='-Gy')
ax.plot(t[:lg],Gyro.flatten()[2:3*lg:3],'g',label='Gz')
plt.legend()
plt.show()
```

```python
dt = np.diff(t);
Gy = Gyro.flatten()[1::3]
ld = min(len(Gy),len(dt))
ang_GyrY = np.cumsum(Gy[:ld]*dt[:ld])

# parts to plot
fig, ax = plt.subplots()
ang = np.arcsin(AccX/np.sqrt(AccX**2 + AccZ**2))
#ax.plot(t,AccX,'b--',lw=2)
#ax.plot(t,AccZ,'g--',lw=1)
ax.plot(t[:la],ang[:la]*57.3,'r',lw=2,label='Ang from Accels')
ax.plot(t[:lg-1],ang_GyrY[:lg-1],'g',lw=1.5,label='Integrated Ang from GyroY')
ax.plot([min(t), max(t)],[90,90],'k:',lw=1,label='+90')
ax.plot([min(t), max(t)],[-90,-90],'k:',lw=1,label='-90')
ax.set_ylim([-100, 100])
plt.legend()
plt.show()
```

## Plot Raw encoder rate compared to filtered values

```python
Fs = 1/np.median(np.diff(t))
print(Fs)

# check timing
plt.figure(2,figsize=(8,5))
ax = plt.subplot(111)
ax.grid(True)
ax.plot(np.diff(t)[500:]-1/Fs,'x')
ax.set_ylabel('Delta Time error')
ax.set_xlabel('Time [s]')
plt.show()
```

```python
try:
    import ahrs
except:
    %pip install ahrs
    # https://ahrs.readthedocs.io/en/latest/    import ahrs
    import ahrs

from ahrs import Quaternion
from ahrs.filters import AngularRate, Tilt, Mahony, Madgwick, EKF

import csv


num_samples = min(len(mAcc),len(Gyro))
II = np.arange(num_samples)
```

```python
angular_rate = AngularRate(gyr=Gyro/57.3,frequency=Fs)

RPY_AR = np.zeros((num_samples, 3))
for ii in range(num_samples):
    RPY_AR[ii] = Quaternion(angular_rate.Q[ii]).to_angles()
```

```python
plt.figure(2,figsize=(8,5))
ax1 = plt.subplot(111)
ax1.grid(True)
ax1.plot(t[:lg],RPY_AR[:,0]*180/np.pi, 'b-',label=r'R')
ax1.plot(t[:lg], RPY_AR[:,1]*180/np.pi, 'r-',label=r'P')
#ax1.plot(t[:lg],RPY_AR[:,2]*180/np.pi, 'g-',label=r'Y')
ax1.grid(True)
ax1.set_title('RPY from Gyros')
ax1.set_xlabel('Time (sec)')
ax1.set_ylabel('RPY (deg)')
ax1.legend(loc='best')
plt.show()
```

```python
orientation_mah = Mahony(gyr=Gyro[:ld,:]/57.3, acc=mAcc[:ld,:], frequency=Fs)
RPY_Mah = np.zeros((num_samples, 3))
for ii in range(num_samples):
    RPY_Mah[ii] = Quaternion(orientation_mah.Q[ii]).to_angles()
```

```python
plt.figure(2,figsize=(8,5))
ax2 = plt.subplot(111)

ax2.plot(t[II],RPY_Mah[II,0]*180/np.pi, 'b-',label=r'R')
ax2.plot(t[II],RPY_Mah[II,1]*180/np.pi, 'r-',label=r'P')
#ax2.plot(t[II],RPY_Mah[II,2]*180/np.pi, 'g-',label=r'Y')
ax2.grid(True)
ax2.set_xlabel('Time (sec)')
ax2.set_ylabel('RPY (deg)')
#ax2.set_ylim([-90, 90])
ax2.legend()
plt.show()
```
